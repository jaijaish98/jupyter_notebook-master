# Linear Regression Algorithm

```python
"""Linear regression algorithm"""

import pandas as pd
import csv
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split 
from sklearn import datasets, linear_model, metrics 
import numpy as np
reader = csv.reader(open("Book3.csv"))
data={}
for row in reader:
    data[row[0]]=row[1:]
train_data= np.vstack((data['Current arrear1'], data['Stand arrear1'],data['CGPA1'],data['Current arrear2'], data['Stand arrear2'],data['CGPA2'],data['Current arrear3'], data['Stand arrear3'],data['CGPA3'],data['Current arrear4'], data['Stand arrear4'],data['CGPA4'],data['Current arrear5'], data['Stand arrear5'],data['CGPA5'],data['Current arrear6'], data['Stand arrear6'],data['CGPA6'],data['Current arrear7'], data['Stand arrear7'],data['CGPA7'])).T
target_data=data['CGPA8']
X_train, X_test, y_train, y_test = train_test_split(train_data,target_data)
reg = linear_model.LinearRegression() 
reg.fit(X_train, y_train) 
X_new = np.array([[0,0,7.15,0,0,6.76,1,1,6.42,0,0,6.82,0,0,7.06,0,0,7.17,0,0,7.14]])
prediction =reg.predict(X_new)
print(prediction)
```

# Decision Tree Algorithm

```python
"""Decision tree algorithm"""

from sklearn.tree import DecisionTreeClassifier
import pandas as pd
import csv
from sklearn.model_selection import train_test_split 
from sklearn import datasets, linear_model, metrics 
import numpy as np
from sklearn.metrics import accuracy_score

reader = csv.reader(open("Book3.csv"))
data={}
for row in reader:
    data[row[0]]=row[1:]
    
train_data= np.vstack((data['Current arrear1'], data['Stand arrear1'],data['CGPA1'],data['Current arrear2'], data['Stand arrear2'],data['CGPA2'],data['Current arrear3'], data['Stand arrear3'],data['CGPA3'],data['Current arrear4'], data['Stand arrear4'],data['CGPA4'],data['Current arrear5'], data['Stand arrear5'],data['CGPA5'],data['Current arrear6'], data['Stand arrear6'],data['CGPA6'],data['Current arrear7'], data['Stand arrear7'],data['CGPA7'])).T
target_data=data['CGPA8']
X_train, X_test, y_train, y_test = train_test_split(train_data,target_data,test_size=0.01)
tree = DecisionTreeClassifier(random_state=0)
tree.fit(X_train, y_train)
X_new = X_test
prediction =tree.predict(X_new)
print(prediction)

decisionTreeAccuracy = accuracy_score(y_train, tree.predict(X_train))
print ("Accuracy :: ", decisionTreeAccuracy)

```

```python
# To support both python 2 and python 3
from __future__ import division, print_function, unicode_literals

# Common imports
import numpy as np
import os

# to make this notebook's output stable across runs
np.random.seed(42)

# To plot pretty figures
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
plt.rcParams['axes.labelsize'] = 14
plt.rcParams['xtick.labelsize'] = 12
plt.rcParams['ytick.labelsize'] = 12
```

```python
student_pass_proba = 0.51
result = (np.random.rand(500, 1) < student_pass_proba).astype(np.int32)
cumulative_result_ratio = np.cumsum(result, axis=0) / np.arange(1, 501).reshape(-1, 1)
```

```python
plt.figure(figsize=(8,3.5))
plt.plot(cumulative_result_ratio)
plt.plot([0, 500], [0.51, 0.51], "k--", linewidth=2, label="51%")
plt.plot([0, 500], [0.5, 0.5], "k-", label="50%")
plt.xlabel("Number of students")
plt.ylabel("pass percent")
plt.legend(loc="lower right")
plt.axis([0, 500, 0.42, 0.58])
plt.show()
```

```python
"""Random Forest algorithm"""

from sklearn.tree import DecisionTreeClassifier
import pandas as pd
import csv
from sklearn.model_selection import train_test_split 
from sklearn import datasets, metrics 
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

import numpy as np
reader = csv.reader(open("Book3.csv"))
data={}
for row in reader:
    data[row[0]]=row[1:]
    
train_data= np.vstack((data['Current arrear1'], data['Stand arrear1'],data['CGPA1'],data['Current arrear2'], data['Stand arrear2'],data['CGPA2'],data['Current arrear3'], data['Stand arrear3'],data['CGPA3'],data['Current arrear4'], data['Stand arrear4'],data['CGPA4'],data['Current arrear5'], data['Stand arrear5'],data['CGPA5'],data['Current arrear6'], data['Stand arrear6'],data['CGPA6'],data['Current arrear7'], data['Stand arrear7'],data['CGPA7'])).T
target_data=data['CGPA8']
X_train, X_test, y_train, y_test = train_test_split(train_data,target_data,test_size=0.01)
tree = RandomForestClassifier(random_state=0)
tree.fit(X_train, y_train)
X_new= X_test
prediction =tree.predict(X_new)
print(prediction)


# Test Accuracy
randomForestAccuracy = accuracy_score(y_train, tree.predict(X_train))
print ("Accuracy :: ", randomForestAccuracy)

```

```python
import matplotlib.pyplot as plt 
  
# portion covered by each label 
model = [randomForestAccuracy,decisionTreeAccuracy ] 
  
# color for each label 
colors = ['g', 'y'] 
activities = ['RandomForest', 'DecisionTree',] 
# plotting the pie chart 
plt.pie(model, labels = activities, colors=colors,  
        startangle=90, shadow = True, explode = (0, 0), 
        radius = 1.2, autopct = '%1.1f%%') 
  
# plotting legend 
plt.legend() 
# showing the plot 
plt.show()
```
